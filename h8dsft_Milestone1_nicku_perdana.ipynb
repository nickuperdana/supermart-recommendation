## D5. Konversi tipe data `order_date`

```python
# konversikan tipe data pada kolom 'order_date' dari string menjadi 'datetime64[ns]'
orderDate = salesRename_df['order_date'] # panggil kolom 'order_date' pada dataframe
newOrderDate = [] # buat list kosong untuk menampung

# buat loop untuk membaca dan memroses tiap-tiap entry data pada kolom 'date' secara berurutan
for entry in orderDate:
    # jalankan kode ini bila separator tanggal adalah '/'
    try:
        month = entry.split('/')[0] # extract string yang menunjukkan info bulan
        date = entry.split('/')[1] # extract string yang menunjukkan info tanggal
        year = entry.split('/')[2] # extract string yang menunjukkan info tahun
        
        # jalankan langkah kondisional bila penulisan tanggal 1 s/d 9 ditulis hanya dengan 1 digit, bukan 01 s/d 09
        if len(date) == 1:
            date = '0' + date # ubah variabel date untuk tertampil sebagai '0x' dari yang semula hanya 'x'
        else:
            pass
        
        # jalankan langkah kondisional bila penulisan bulan 1 s/d 9 ditulis hanya dengan 1 digit, bukan 01 s/d 09
        if len(month) == 1:
            month = '0' + month # ubah variabel bulan untuk tertampil sebagai '0x' dari yang semula hanya 'x'
        else:
            pass
        
        newDateEntry = f'{year}/{month}/{date}' # rangkai susunan tanggal sesuai standar
        newOrderDate.append(newDateEntry) # masukkan hasil rangkaian tanggal yang sudah sesuai ke dalam list kosong
        
    # jalankan kode ini bila separator tanggal adalah '-'
    except:
        month = entry.split('-')[0] # extract string yang menunjukkan info bulan
        date = entry.split('-')[1] # extract string yang menunjukkan info tanggal
        year = entry.split('-')[2] # extract string yang menunjukkan info tahun
        
        # jalankan langkah kondisional bila penulisan tanggal 1 s/d 9 ditulis hanya dengan 1 digit, bukan 01 s/d 09
        if len(date) == 1:
            date = '0' + date # ubah variabel date untuk tertampil sebagai '0x' dari yang semula hanya 'x'
        else:
            pass
        
        # jalankan langkah kondisional bila penulisan bulan 1 s/d 9 ditulis hanya dengan 1 digit, bukan 01 s/d 09
        if len(month) == 1:
            month = '0' + month # ubah variabel bulan untuk tertampil sebagai '0x' dari yang semula hanya 'x'
        else:
            pass
        
        newDateEntry = f'{year}/{month}/{date}' # rangkai susunan tanggal sesuai standar
        newOrderDate.append(newDateEntry) # masukkan hasil rangkaian tanggal yang sudah sesuai ke dalam list kosong
        
# ganti isi kolom 'order_date' dengan isi yang sudah sesuai dengan standar
salesCleanedDate_df = salesRename_df.copy()
salesCleanedDate_df['order_date'] = newOrderDate

salesCleanedDate_df.head()
```

```python
# konversi tipe data dalam kolom 'order_date'
salesCleanedDate_df['order_date'] = pd.to_datetime(salesCleanedDate_df['order_date'], format='%Y/%m/%d') 

salesCleanedDate_df.info()
```

Deskripsi: Berdasarkan hasil dari langkah konversi tipe data tanggal, langkah tersebut sudah berhasil dilakukan, dibuktikan dengan tipe data kolom `order_date` yang sudah berubah menjadi 'datetime64[ns]' tanpa memunculkan nilai null.

## E3. Analisis terhadap Pola Kecenderungan Angka Penjualan (_Sales_) 

```python
# Data sales

columnSales = salesClean_df['sales'] # panggil kolom `sales`

meanSales = columnSales.mean() # hitung nilai rata-rata
medianSales = columnSales.median() # hitung nilai median
modeSales = columnSales.mode()[0] # hitung nilai modus
stdevSales = columnSales.std() # hitung nilai standar deviasi
minSales = columnSales.min() # hitung nilai paling kecil
maxSales = columnSales.max() # hitung nilai paling besar
skewSales = columnSales.skew() # hitung nilai skewness
kurtSales = columnSales.kurt() # hitung nilai kurtosis

print(f'Rata-rata nilai penjualan dalam 1 transaksi: {meanSales} INR.')
print(f'Nilai tengah/median nilai penjualan: {medianSales} INR.')
print(f'Total nilai penjualan yang paling sering muncul dalam 1 transaksi: {modeSales} INR.')
print(f'Standar deviasi angka penjualan: {stdevSales} INR.')
print(f'Nilai penjualan minimum: {minSales} INR.')
print(f'Nilai penjualan maksimuml: {maxSales} INR.')
print(f'Skewness nilai penjualan: {skewSales}.')
print(f'Kurtosis nilai penjualan: {kurtSales}.')
```

Deskripsi:  
- Diketahui nilai penjualan dalam satu kali transaksi ialah sekitar 1496 INR, dengan nilai tengah ada di posisi 1498 item barang. Nilai modus dari angka penjualan sekali transaksi ada di angka 1259 INR.
- Melihat nilai rata-rata dan median yang cenderung berdekatan, serta nilai skewness yang mendekati 0, pola distrbusi angka penjualan bisa jadi memang berdistribusi normal, atau bisa jadi uniform. Hal ini dapat lebih lanjut dilihat dengan menggunakan visualisasi histogram.

Deskripsi: Berdasarkan visualisasi di atas, distribusi nilai penjualan dalam sekali transaksi cenderung seragam/uniform. Artinya, mulai dari angka penjualan paling sedikit hingga paling banyak, variasi besaran nilai penjualan sekali transaksi yang terjadi cenderung tidak banyak berbeda satu sama lain.

```python
# visualisasikan boxplot data kolom `profit`
boxPlotSales = sns.boxplot(x=columnSales).set(xlabel = 'Penjualan (INR)', title = 'Distribusi Angka Penjualan')
```

Deskripsi: Mengingat nilai rerata dan median sales cenderung berdekatan, hal ini membuat kecenderungan data seakan-akan simetris. Hal ini membuat grafik distribusi angka penjualan cenderung memiliki garis median yang ada di tengah grafik. Terlihat pula grafik tidak memunculkan data-data yang terindikasi sebagai outlier. Untuk memastikan temuan tersebut, perlu dilakukan cek dengan menggunakan aturan Z-score, mengingat skewness data ditemukan lebih kecil dari +0,5.

```python
# cek berapa nilai outlier dengan aturan Z-score karena skewness < 0,05

q1Sales = columnSales.quantile(0.25)
q3Sales = columnSales.quantile(0.75)

# Aturan Z-score:
lowBound = meanSales - (3 * stdevSales) # temukan angka batas bawah outlier
upBound = meanSales + (3 * stdevSales) # temukan angka batas atas outlier

lenOriginal = len(salesClean_df)

salesOutlier = salesClean_df[(columnSales <= lowBound) | (columnSales >= upBound)] # Simpan tabel dengan data yang terindikasi outlier
salesNoOutlier = salesClean_df[(columnSales > lowBound) & (columnSales < upBound)] # Simpan tabel dengan data yang bersih dari terindikasi outlier

print(f'Nilai batas bawah penjualan: {lowBound} INR.')
print(f'Nilai batas atas penjualan: {upBound} INR.')
print(f'Jumlah data sebelum disaring dengan kriteria deteksi outlier: {lenOriginal} data.')
print(f'Jumlah data outlier: {len(salesOutlier)} Data ({(len(salesOutlier) * 100)/lenOriginal:.2f}%).')
```

Deskripsi:
- Sebuah data dikatakan sebagai outlier ketika pada kolom sales, terdapat data yang berada di posisi sebagaimana berikut: `-236.08` < `data sales` < `3229,27`

```python
# buat list kosong untuk menampung daftar keluaran info kota yang memenuhi target tujuan bisnis 
cityResult = []
condition = ' (memenuhi kriteria tujuan SMART)'

print('Rerata keuntungan per transaksi di tiap Kota Negara Bagian Tamil Nadu: \n')

# jalankan untuk setiap daftar nama-nama kota yang unik
for city in list(cityUnique):
    # hitung rerata profit penjualan per kota dalam empat tahun
    cityProfitMean = salesClean_df[salesClean_df['city'] == city]['profit'].mean()
    cityProfitStd = salesClean_df[salesClean_df['city'] == city]['profit'].std()
    # hitung banyaknya data penjualan yang terjadi di kota tersebut 
    population = len(salesClean_df[salesClean_df['city'] == city])
    
    # hitung confidence interval untuk rerata profit di kota tersebut
    lowLimit, upLimit = stats.norm.interval(confidence=0.95, loc=cityProfitMean, scale=cityProfitStd/np.sqrt(population))
    # tampilkan detail info batas bawah dan atas rerata profit di kota tersebut beserta statusnya
    outputCity = f'Kota {city}: Low={lowLimit:.2f} INR, High={upLimit:.2f} INR{condition if (meanProfit*1.05) < upLimit else ""}.'
    print(outputCity)
    # simpan info ke dalam list yang telah dibuat sebelumnya
    cityResult.append(outputCity)
    
# lakukan fungsi menghitung daftar kota yang memenuhi kondisi
countCity = 0
for item in cityResult:
    countCity += 1 if condition in item else 0

# tampilkan banyaknya kota yang memenuhi kondisi
print(f'\nBanyaknya kota yang dapat dijajaki bisnis berdasarkan tujuan: {countCity} kota dari sebanyak {len(cityUnique)} di Tamil Nadu.')

```

## E7. Hubungan antara Profit dengan Angka Penjualan per Transaksi  
  
Umumnya, pebisnis mengejar peningkatan angka penjualan sehingga profitabilitas dapat naik sedemikian rupa. Asumsi ini secara mentah cukup tergambar melalui visualisasi yang menggambarkan kenaikan profit ada hubungannya dengan kenaikan angka penjualan per transaksi yang dilakukan oleh pelanggan.

Dalam menjawab asumsi ini, analisis korelasi dapat dilakukan dengan mengujihubungkan dua variabel (profit dan sales) dengan menggunakan hipotesis sebagaimana berikut:

> H0: Tidak ada hubungan yang signifikan antara nilai profit dan angka penjualan  
> H1: Ada hubungan yang signifikan antara nilai profit dan angka penjualan
  
Dalam menguji hipotesis ini, uji Spearman's Rho dapat dipilih dengan pertimbangan:
- Kedua variabel memuat data numerikal continuous yang mana salah satu variabelnya (profit) memiliki nilai outlier.
- Spearman's Rho menjadi pilihan yang tepat untuk menguji variabel dengan nilai outlier untuk mereduksi bias dalam pengukuran uji korelasi

```python
# Lakukan uji korelasi spearman terhadap variabel profit dan sales
correlationRho, pValueRho = stats.spearmanr(salesClean_df['profit'], salesClean_df['sales'])

print(f'Rho Correlation: {correlationRho}, p-value: {pValueRho}\n')

# buat visualisasi sebaran data profit dan sales dengan scatterplot
sns.lmplot(data=salesClean_df, x='sales', y='profit', scatter_kws={"s":11}).set(title='Hubungan antara Angka Keuntungan dan Penjualan\n Toko Supermart', xlabel='Penjualan', ylabel='Keuntungan')
plt.subplots_adjust(top=1.2)
plt.show()
```

Deskripsi: Berdasarkan uji korelasi Spearman's terhadap variabel profit dan sales, diketahui nilai p-value dari korelasi ini lebih kecil dari nilai p-value konsensus (p-value: 0.0). Hal ini menunjukkan bahwa H0 berhasil untuk ditolak. Dengan demikian, angka profit dan penjualan/sales memiliki hubungan yang signifikan tergambar sebagai hubungan positif. Kenaikan angka profit cenderung bersamaan dengan kenaikan angka sales.

## E8. Perbedaan Profit di Tiap-tiap kota

Untuk melakukan uji perbedaan nilai profit yang didapatkan di masing-masing kota, perlu dilakukan uji statistik inferensial. Dalam kasus ini, tes ANOVA dapat dilakukan dengan memertimbangkan sebagaimana berikut:
- Nama-nama kota diperlakukan sebagai kategori
- rerata profit diperlakukan sebagai data numerikal _continuous_. 
  
Dalam melakukan uji statistik ANOVA, ada pula hipotesisnya ialah sebagaimana berikut:
> H0: Tidak ada perbedaan profit per transaksi yang signifikan di antara beberapa kategori kota  
> H1: Ada perbedaan profit per transaksi yang signifikan di antara beberapa kategori kota


```python
# panggil data profit untuk kolom kota masing-masing
vellore = salesClean_df['profit'][salesClean_df['city'] == 'Vellore']
krishnagiri = salesClean_df['profit'][salesClean_df['city'] == 'Krishnagiri']
perambalur = salesClean_df['profit'][salesClean_df['city'] == 'Perambalur']
dharmapuri = salesClean_df['profit'][salesClean_df['city'] == 'Dharmapuri']
ooty = salesClean_df['profit'][salesClean_df['city'] == 'Ooty']
trichy = salesClean_df['profit'][salesClean_df['city'] == 'Trichy']
ramanadhapuram = salesClean_df['profit'][salesClean_df['city'] == 'Ramanadhapuram']
tirunelveli = salesClean_df['profit'][salesClean_df['city'] == 'Tirunelveli']
chennai = salesClean_df['profit'][salesClean_df['city'] == 'Chennai']
karur = salesClean_df['profit'][salesClean_df['city'] == 'Karur']
namakkal = salesClean_df['profit'][salesClean_df['city'] == 'Namakkal']
dindigul = salesClean_df['profit'][salesClean_df['city'] == 'Dindigul']
kanyakumari = salesClean_df['profit'][salesClean_df['city'] == 'Kanyakumari']
bodi = salesClean_df['profit'][salesClean_df['city'] == 'Bodi']
tenkasi = salesClean_df['profit'][salesClean_df['city'] == 'Tenkasi']
viluppuram = salesClean_df['profit'][salesClean_df['city'] == 'Viluppuram']
madurai = salesClean_df['profit'][salesClean_df['city'] == 'Madurai']
salem = salesClean_df['profit'][salesClean_df['city'] == 'Salem']
cumbum = salesClean_df['profit'][salesClean_df['city'] == 'Cumbum']
nagercoil = salesClean_df['profit'][salesClean_df['city'] == 'Nagercoil']
pudukottai = salesClean_df['profit'][salesClean_df['city'] == 'Pudukottai']
theni = salesClean_df['profit'][salesClean_df['city'] == 'Theni']
coimbatore = salesClean_df['profit'][salesClean_df['city'] == 'Coimbatore']
virudhunagar = salesClean_df['profit'][salesClean_df['city'] == 'Virudhunagar']

# lakukan uji anova
cityAnova = stats.f_oneway(vellore, krishnagiri, perambalur, dharmapuri, ooty,
                           trichy, ramanadhapuram, tirunelveli, chennai, karur,
                           namakkal, dindigul, kanyakumari, bodi, tenkasi,
                           viluppuram, madurai, salem, cumbum, nagercoil,
                           pudukottai, theni, coimbatore, virudhunagar)

# tampilkan hasilnya
print('P-value:', cityAnova.pvalue)
```

Deskripsi: Terlihat bahwa p-value untuk uji Anova berikut berada di posisi lebih dari nilai kritis sebesar 0,05. Berdasarkan hasil uji Anova yang mengkategorikan pendapatan profit per transaksi di beberapa kota Supermart beroperasi, terlihat bahwa antar kota tidak memiliki perbedaan profit per transaksi yang signifikan. Artinya, seluruh kota menerapkan aturan margin profit yang setara dan cenderung seragam, dan tidak ada kota yang terindikasi menerapkan aturan/margin profit yang berbeda.  
  
Dengan demikian, parameter lokasi untuk investasi franchise dapat tetap terfokus pada kota-kota dengan rerata profit yang ada di area 5% lebih tinggi dari rerata profit secara keseluruhan, tanpa fokus pada konfigurasi margin profit di masing-masing kota. ## E9. Perbedaan Profit per Transaksi Berdasarkan Kategori Barang Jualan

Untuk mengetahui signifikansi perbedaan profit per transaksi yang dikelompokkan berdasarkan nama-nama kategori barang jualan, tes Anova dapat kembali dilakukan dengan memperlakukan nama kategori barang sebagai kelompok kategorikal untuk tes Anova yang akan kita lakukan.  
  
Ada pula hipotesis untuk tes Anova untuk nama kategori barang jualan terhadap profit per transaksi ialah sebagaimana berikut:
> H0: Tidak ada perbedaan profit per transaksi yang signifikan di antara beberapa nama kategori produk yang dijual  
> H1: Ada perbedaan profit per transaksi yang signifikan di antara beberapa nama kategori produk yang dijual

```python
# panggil data profit untuk kategori produk masing-masing
oilMasala = salesClean_df['profit'][salesClean_df['category'] == 'Oil & Masala']
beverages = salesClean_df['profit'][salesClean_df['category'] == 'Beverages']
foodGrains = salesClean_df['profit'][salesClean_df['category'] == 'Food Grains']
fruitVeggies = salesClean_df['profit'][salesClean_df['category'] == 'Fruits & Veggies']
bakery = salesClean_df['profit'][salesClean_df['category'] == 'Bakery']
snacks = salesClean_df['profit'][salesClean_df['category'] == 'Snacks']
eggsMeatFish = salesClean_df['profit'][salesClean_df['category'] == 'Eggs, Meat & Fish']
# lakukan uji anova
categoryAnova = stats.f_oneway(oilMasala, beverages, foodGrains,
                           fruitVeggies, bakery, snacks, eggsMeatFish)

# tampilkan hasilnya
print('P-value:', categoryAnova.pvalue)
```

Deskripsi: Berdasarkan uji statistik ANOVA terkait perbedaan profit per transaksi terhadap beberapa kategori barang, diketahui p-value yang didapat ialah sebesar 0,78. Kondisi p-value yang lebih besar dibandingkan nilai kritis 0,05 membuat H0 gagal untuk ditolak. Artinya, serangkaian angka profit per transaksi di masing-masing kategori memiliki nilai yang secara signifikan tidak berbeda satu sama lain. Hal ini menjadikan adanya kemungkinan, konsumen membeli barang dengan satuan secara konsisten sehingga rerata profit yang didapat cenderung relatif sama satu sama lain. Dalam hal ini, bisnis Supermart yang akan dijajaki dapat memperlakukan seluruh kategori barang secara setara dengan menjual seluruh kategori di seluruh gerai di Tamil Nadu, India
